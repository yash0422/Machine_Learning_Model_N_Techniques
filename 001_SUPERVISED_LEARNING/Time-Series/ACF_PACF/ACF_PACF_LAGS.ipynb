#### IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rc('xtick', labelsize = 30)
matplotlib.rc('ytick', labelsize = 30)

import seaborn as sns
sns.set(style = "whitegrid", color_codes = True)

from statsmodels.tsa.stattools import acf, pacf
```

#### IMPORT DATASET

```python
WMT_Stock = pd.read_csv('walmart_stock.csv')
```

```python
WMT_Stock.head()
```

```python
# DROP UNWANTED COLUMNS
WMT_Stock = WMT_Stock.drop(['Open', 'High', 'Low', 'Volume', 'Adj Close'], axis = 1)
```

```python
WMT_Stock.columns
```

#### CONVERT DATE COLUMN TO to_datetime FORMAT

```python
WMT_Stock['Date'] = pd.to_datetime(WMT_Stock['Date'])
```

#### SET DATE COLUMN AS INDEX

```python
WMT_Stock.set_index('Date', inplace = True)
```

```python
WMT_Stock.head()
```

#### PLOT THE DATASET

```python
WMT_Stock.plot()
```

#### CREATE ACF PLOT TO CHECK CORRELATION OF LAGS

ACF PLOT will show us till how many lags there is a CORRELATION between out Data point in Close Column.

```python
y = WMT_Stock['Close']
```

```python
# Calling AUTO CORRELATION FUNCTION
ACF_lag = acf(y, nlags = 500)

plt.figure(figsize = (16, 7))
plt.plot(ACF_lag, marker = '+')
plt.axhline(y = 0, linestyle = '--', color = 'green')
plt.axhline(y = -1.96 / np.sqrt(len(y)), linestyle = '-.', color = 'red')
plt.axhline(y = 1.96 / np.sqrt(len(y)), linestyle = '-.', color = 'blue')
plt.title('AUTOCORRELATION')
plt.xlabel('NUMBER_OF_LAGS')
plt.ylabel('CORRELATION')
plt.tight_layout()
```

```python
# Calling AUTO CORRELATION FUNCTION
ACF_lag = acf(y, nlags = 400)

plt.figure(figsize = (16, 7))
plt.plot(ACF_lag, marker = '+')
plt.axhline(y = 0, linestyle = '--', color = 'green')
plt.axhline(y = -1.96 / np.sqrt(len(y)), linestyle = '-.', color = 'red')
plt.axhline(y = 1.96 / np.sqrt(len(y)), linestyle = '-.', color = 'blue')
plt.title('AUTOCORRELATION')
plt.xlabel('NUMBER_OF_LAGS')
plt.ylabel('CORRELATION')
plt.tight_layout()
```

```python
# Calling AUTO CORRELATION FUNCTION
ACF_lag = acf(y, nlags = 300)

plt.figure(figsize = (16, 7))
plt.plot(ACF_lag, marker = '+')
plt.axhline(y = 0, linestyle = '--', color = 'green')
plt.axhline(y = -1.96 / np.sqrt(len(y)), linestyle = '-.', color = 'red')
plt.axhline(y = 1.96 / np.sqrt(len(y)), linestyle = '-.', color = 'blue')
plt.title('AUTOCORRELATION')
plt.xlabel('NUMBER_OF_LAGS')
plt.ylabel('CORRELATION')
plt.tight_layout()
```

```python
# Calling AUTO CORRELATION FUNCTION
PACF_lag = pacf(y, nlags = 20, method = 'ols')

plt.figure(figsize = (16, 7))
plt.plot(PACF_lag, marker = '+')
plt.axhline(y = 0, linestyle = '--', color = 'green')
plt.axhline(y = -1.96 / np.sqrt(len(y)), linestyle = '-.', color = 'red')
plt.axhline(y = 1.96 / np.sqrt(len(y)), linestyle = '-.', color = 'blue')
plt.title('PARTIAL_AUTOCORRELATION')
plt.xlabel('NUMBER_OF_LAGS')
plt.ylabel('CORRELATION')
plt.tight_layout()
```
