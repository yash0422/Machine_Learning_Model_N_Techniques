```python
marketing_df = marketing_df.drop(['Customer', 'Vehicle_Class', 'avg_vehicle_age', 'months_last_claim', 
                                  'Total_Claim_Amount'], axis = 1)
```

```python
marketing_df.head()
```

```python
mark_array=marketing_df.values
```

```python
mark_array[:, 1] = mark_array[:, 1].astype(float)
mark_array[:, 3] = mark_array[:, 3].astype(float)
mark_array[:, 5] = mark_array[:, 5].astype(float)
mark_array[:, 6] = mark_array[:, 6].astype(float)
```

```python
pd.DataFrame(mark_array)
```

```python
'Huang', 'Cao', 'random'
```

```python
kproto_cao = KPrototypes(n_clusters = 3, init ='Cao', verbose = 2, max_iter = 100)
clusters_cao = kproto_cao.fit_predict(mark_array, categorical=[0, 2, 4])
```

```python
print('- - - - - - - - - - - - - - - - - - - - ')
print('Cluser Centroid: ', kproto_cao.cluster_centroids_)
print('- - - - - - - - - - - - - - - - - - - - ')
print('Coa Cost: ', kproto_cao.cost_)
print('- - - - - - - - - - - - - - - - - - - - ')
print('Epoch Cost: ', kproto_cao.epoch_costs_)
print('- - - - - - - - - - - - - - - - - - - - ')
print('Gamma: ', kproto_cao.gamma)
print('- - - - - - - - - - - - - - - - - - - - ')
print('Cao Labels: ', kproto_cao.labels_)
print('- - - - - - - - - - - - - - - - - - - - ')
print('Max Iteration: ', kproto_cao.max_iter)
print('- - - - - - - - - - - - - - - - - - - - ')
print('N_Clusters: ', kproto_cao.n_clusters)
print('- - - - - - - - - - - - - - - - - - - - ')
print('N_Iteration: ', kproto_cao.n_iter_)
print('- - - - - - - - - - - - - - - - - - - - ')
print('N_Jobs: ', kproto_cao.n_jobs)
print('- - - - - - - - - - - - - - - - - - - - ')
print('Num_Dissim: ', kproto_cao.num_dissim)
```

```python
kproto_huang = KPrototypes(n_clusters = 3, init ='Huang', verbose = 2, max_iter = 100)
clusters_huang = kproto_huang.fit_predict(mark_array, categorical=[0, 2, 4])
```

```python
kproto_random = KPrototypes(n_clusters = 3, init ='random', verbose = 2, max_iter = 100)
clusters_random = kproto_random.fit_predict(mark_array, categorical=[0, 2, 4])
```
