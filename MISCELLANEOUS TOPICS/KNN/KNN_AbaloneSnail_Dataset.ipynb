```python
y = abalone["Rings"]
X = abalone.drop(columns = "Rings")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initializing classifier and giving hyperparameter k=5
KNC = KNeighborsClassifier(n_neighbors = 5)

# training classifier & fitting Training Set
model_kNeighborsClassifier = KNC.fit(X_train, y_train)

# Predicting on test dataset
pred_knc = model_kNeighborsClassifier.predict(X_test)

# Evaluation of KNeighborsClassifier 
acc_knc = accuracy_score(y_test, pred_knc) # evaluating accuracy score
print('accuracy score of KNeighbors Classifier is:', acc_knc * 100)
```

#### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Initializing classifier with one-v-rest approach.
# random_state is # to ensure same results in every execution.
LOG_REG = LogisticRegression(multi_class = 'ovr', random_state = 3)

# training classifier
model_Log_Reg = LOG_REG.fit(X_train, y_train)

# Predicting on test dataset
pred_Log_Reg = model_Log_Reg.predict(X_test)

# Evaluation of KNeighborsClassifier 
acc_log_reg = accuracy_score(y_test, pred_Log_Reg) # evaluating accuracy score
print('accuracy score of LogisticRegression Classifier is:', acc_log_reg * 100)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initializing classifier with one-v-rest approach.
# random_state is # to ensure same results in every execution.
DTC = DecisionTreeClassifier(random_state = 0, max_depth = 3)

# training classifier
model_Dec_Tre = DTC.fit(X_train, y_train)

# Predicting on test dataset
pred_Dec_Tre = model_Dec_Tre.predict(X_test)

# Evaluation of KNeighborsClassifier 
acc_Dec_Tre = accuracy_score(y_test, pred_Dec_Tre) # evaluating accuracy score
print('accuracy score of DecisionTree Classifier is:', acc_Dec_Tre * 100)
```

```python
# Initializing classifier. random_state is to ensure same results in every execution. max_depth is to specify number of splits
dt = DecisionTreeClassifier(random_state=0, max_depth = 3)
# training classifier
dt.fit(X_train, y_train)
# Evaluate the classifier
print(dt.score(X_test, y_test))
```
